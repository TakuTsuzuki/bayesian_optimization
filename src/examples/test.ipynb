```python
%matplotlib inline
import sys
import numpy as np
import matplotlib.pyplot as plt

#sys.path.append( '/Users/tsuzuki/BayesianOptimization/bayes_opt')
#from bayesian_optimization import BayesianOptimization
#from helpers import UtilityFunction
from bayes_opt import BayesianOptimization
```

## Experiments

### convex

```python
n_sample=3
n_iter = 10

# results
results = {}
for k in range(1,6):
    results[k]=np.array([])

for i in range(n_sample):
    print (i) 
    #Init funcs
    bo1 = BayesianOptimization(lambda x1: -(x1**2), 
                               {'x1': (-4, 4)})
    bo2 = BayesianOptimization(lambda x1,x2: -(x1**2+x2**2), 
                               {'x1': (-4, 4),'x2': (-4, 4)})
    bo3 = BayesianOptimization(lambda x1,x2,x3: -(x1**2+x2**2+x3**2), 
                               {'x1': (-4, 4),'x2': (-4, 4),'x3': (-4, 4)})
    bo4 = BayesianOptimization(lambda x1,x2,x3,x4: -(x1**2+x2**2+x3**2+x4**2), 
                               {'x1': (-4, 4),'x2': (-4, 4),'x3': (-4, 4),'x4': (-4, 4)})
    bo5 = BayesianOptimization(lambda x1,x2,x3,x4, x5: -(x1**2+x2**2+x3**2+x4**2+x5**2), 
                               {'x1': (-4, 4),'x2': (-4, 4),'x3': (-4, 4),'x4': (-4, 4),'x5': (-4, 4)})
    
    for j in range(n_iter):
        # calc next experiment
        bo1.maximize(init_points=1, n_iter=1, acq='mi', kappa=3.29)
        bo2.maximize(init_points=1, n_iter=1, acq='mi', kappa=3.29)
        bo3.maximize(init_points=1, n_iter=1, acq='mi', kappa=3.29)
        bo4.maximize(init_points=1, n_iter=1, acq='mi', kappa=3.29)
        bo5.maximize(init_points=1, n_iter=1, acq='mi', kappa=3.29)
        
        # append max 
        results[1] = np.append(results[1],[np.max(bo1.res["all"]["values"])])
        results[2] = np.append(results[2],[np.max(bo2.res["all"]["values"])])
        results[3] = np.append(results[3],[np.max(bo3.res["all"]["values"])])
        results[4] = np.append(results[4],[np.max(bo4.res["all"]["values"])])
        results[5] = np.append(results[5],[np.max(bo5.res["all"]["values"])])
        
results[1]=results[1].reshape((n_sample, n_iter))
results[2]=results[2].reshape((n_sample, n_iter))
results[3]=results[3].reshape((n_sample, n_iter))
results[4]=results[4].reshape((n_sample, n_iter))
results[5]=results[5].reshape((n_sample, n_iter))
```

#### Plot results

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)

ax.plot(-np.mean(results[1],axis=0), c="red", marker="^", label="dim1")
ax.plot(-np.mean(results[2],axis=0), c="blue", marker="^", label="dim2")
ax.plot(-np.mean(results[3],axis=0), c="yellow", marker="^", label="dim3")
ax.plot(-np.mean(results[4],axis=0), c="black", marker="^", label="dim4")
ax.plot(-np.mean(results[5],axis=0), c="green", marker="^", label="dim5")

ax.set_ylim(0,40)

ax.set_xlabel("query", size=12)
ax.set_ylabel("value(max)", size=12)

ax.grid(True)
ax.legend(loc='upper right')

fig.show()
```

### non-convex
